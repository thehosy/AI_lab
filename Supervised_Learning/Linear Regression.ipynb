# Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(11)
```

```python
X = np.random.rand(1000, 1)
Y = 2 * X + 5 + .2 * np.random.randn(1000, 1)

one = np.ones_like(X)
X_bar = np.concatenate((one, X), axis = 1)

Ap = X_bar.T @ X_bar
b = X_bar.T @ Y

w = np.linalg.pinv(Ap) @ b

x0 = np.linspace(0, 1, 2)
y0 = w[0, 0] + w[1, 0] * x0

print(w.T)

plt.figure(figsize=(12,8), dpi= 100, facecolor='w', edgecolor='k')

plt.plot(X, Y, 'b.')
plt.plot(x0, y0, 'r', linewidth = 1)


plt.show()
```

```python
X.shape
```

```python
n = 50
N = 1000

plt.figure(figsize=(18, 9), dpi= 100, facecolor='w', edgecolor='k')

def f(x,y):
    return np.array([[0.5 / N * np.linalg.norm(Y - X_bar @ np.array([[x[i, j]], [y[i, j]]])) ** 2 for j in range(n)] for i in range(n)] )

x_ = np.linspace(2, 8, n)
y_ = np.linspace(0, 4, n)

x,y = np.meshgrid(x_, y_)

levels = f(x, y)

plt.plot(5, 2, 'ro')

plt.contour(x, y, levels, 100)
plt.colorbar()
plt.show()
```
